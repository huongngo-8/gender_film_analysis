```python
# removing leading and trailing characters from columns with str type
str_cols = list(netflix_titles.columns)
str_cols.remove('release_year')

for i in str_cols:
    netflix_titles[i] = netflix_titles[i].str.strip()
```

```python
# examining rows with null values for date_added column
rows = []
for i in range(len(netflix_titles)):
    if netflix_titles['date_added'].iloc[i] == "":
        rows.append(i)
    
# examine those rows to confirm null state
netflix_titles.loc[rows, :]
```

```python
# extracting months added and years added
month_added = []
year_added = []
for i in range(len(netflix_titles)):
    if i in rows:
        month_added.append(0)
        year_added.append(0)
    else:
        date = netflix_titles['date_added'].iloc[i].split(" ")
        month_added.append(date[0])
        year_added.append(int(date[2]))

for i, month in enumerate(month_added):
    if month != 0:
        datetime_obj = datetime.strptime(month, "%B")
        month_number = datetime_obj.month
        month_added[i] = month_number
        
# checking all months
print(set(month_added))
print(set(year_added))

netflix_titles.insert(7, "month_added", month_added, allow_duplicates = True)
netflix_titles.insert(8, "year_added", year_added, allow_duplicates = True)

netflix_titles.head()
```

```python
# get length of movie or number of seasons of show
def getDuration(data):
    count = 0
    durations = []
    for value in data:
        # filling in missing values
        if type(value) is float:
            durations.append(0)
        else:
            values = value.split(" ")
            durations.append(int(values[0]))
    return durations
```

```python
# function to get unique values of a column
def getUnique(data):
    unique_values = set()
    for value in data:
        if type(value) is float:
            unique_values.add(None)
        else:
            values = value.split(", ")
            for i in values:
                unique_values.add(i)
    return list(unique_values)
```

```python
def guessGender(name, det):
    first_name = name.split(" ")[0]
    if det.get_gender(first_name) == "male":
        gender = "male"
    elif det.get_gender(first_name) == "female":
        gender = "female"
    else:
        gender = "unknown"
    return gender
```

```python
# getting the gender distribution of casts
professions_dict = dict(zip(imdb_names['primaryName'], imdb_names['primaryProfession']))
genders = [''] * len(netflix_rating)
det = gen.Detector()
for i, row in netflix_rating.iterrows():
    if row['cast'] != "":
        cast = row['cast']
        temp = []

        for name in cast.split(", "):
            gender = ""
            if (name in professions_dict.keys()):
                if type(professions_dict[name]) is not float:
                    professions = professions_dict[name]
                    if ("actor" in professions):
                        gender = "male"
                    elif ("actress" in professions):
                        gender = "female"
                    else:
                        gender = guessGender(name, det)
                else:
                    gender = guessGender(name, det)
            else:
                gender = guessGender(name, det)
            temp.append(gender)
        genders[i] = temp
cast_genders = genders
cast_genders
```

```python
cast_majority = [''] * len(netflix_rating)
for i, genders in enumerate(cast_genders):
    majority = ''
    if genders != '':
        dist = dict(Counter(genders))
        if len(list(set(list(dist.values())))) == 1:
            majority = 'equal'
        else:
            majority = max(dist, key = dist.get)
            if majority == "unknown":
                majority = 'none'
        cast_majority[i] = majority
    else:
        cast_majority[i] = 'none' 
cast_majority
```

```python
cast_ratio = [''] * len(netflix_rating)
for i, genders in enumerate(cast_genders):
    ratio = 0
    if genders != '':
        dist = dict(Counter(genders))
        if "female" in dist.keys():
            ratio = dist["female"] / sum(dist.values())
    cast_ratio[i] = ratio
cast_ratio
```

```python
# getting the genders of the directors
det = gen.Detector()
director_genders = [''] * len(netflix_rating)
for i, directors in enumerate(netflix_rating['director']):
    if directors != "":
        temp = []
        directors_list = directors.split(", ")
        for director in directors_list:
            gender = guessGender(director, det)
            temp.append(gender)
        director_genders[i] = temp
director_genders
```

```python
director_majority = [''] * len(netflix_rating)
for i, genders in enumerate(director_genders):
    majority = ''
    if genders != '':
        dist = dict(Counter(genders))
        if len(list(set(list(dist.values())))) == 1 and len(list(dist.keys())) != 1:
            majority = 'equal'
        else:
            majority = max(dist, key = dist.get)
            if majority == "unknown":
                majority = 'none'
        director_majority[i] = majority
    else:
        director_majority[i] = 'none' 
director_majority
```

```python
temp_df = pd.DataFrame(list(zip(cast_genders, cast_majority, cast_ratio, director_genders, director_majority)), columns = ['cast_genders', 'cast_majority', 'cast_ratio', 'director_genders', 'director_majority'])
netflix_rating = pd.concat([netflix_rating, temp_df], axis = 1)
netflix_rating.head(5)
```
